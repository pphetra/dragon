```python
import folium
```

```python
columbo_coord = [6.91376, 79.8752]
m = folium.Map(location=columbo_coord, zoom_start=12)

m
```

```python
import pandas as pd

df = pd.read_csv("fake_risk.csv", )
```

```python
import json
from folium import plugins

geopath = json.load(open(r'CMCwards_wgs84.geojson'))

def render(feature):
    ward_no = feature.properties['WD_NUMBER']
    risk = df.loc[ward_no].risk
    return {
        'fillColor': '#ffaf00', 
        'color': 'blue'
    }

g = folium.GeoJson(geopath,
          overlay=True,
          style_function=render)
g.add_to(m)
m
           
```

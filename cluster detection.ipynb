```python
import pandas as pd
import xlrd
```

```python
df = pd.read_excel('2014 DDB.xlsx', usecols=['Date of Onset', 'N', 'E', 'Ward \nNo.'], names=['date', 'lat', 'lng', 'ward'])
```

```python
df.head()
```

```python
valid_df = df[df['lat'] > 0]
```

```python
valid_df.head()
```

```python
valid_df.count()
```

```python
import folium
```

```python
m = folium.Map(location=[6.9271, 79.8612])
```

```python
folium.CircleMarker(location=[6.91376, 79.8752]).add_to(m)
m
```

```python
valid_df.apply(lambda row:folium.CircleMarker(
    location=[row["lat"], row["lng"]],
    color='red',
    radius=1).add_to(m),
         axis=1, ignore_failures=True)
```
